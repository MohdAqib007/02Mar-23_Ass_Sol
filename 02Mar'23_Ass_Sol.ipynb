```python
#Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
'''import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.'''

#Ans-

'''A scatter plot is a type of graph that displays the relationship between two variables in a two-dimensional coordinate system. 
It is often used to investigate the correlation or association between two variables.

In a scatter plot, each point on the graph represents an observation in the data, with the x-axis representing one variable and the y-axis representing the other variable. 
The position of each point is determined by the values of the two variables for that observation.

The scatter plot is a useful tool for visualizing the pattern and strength of the relationship between the variables. 
If the points on the graph are tightly clustered around a line or curve, this indicates a strong correlation between the variables. 
On the other hand, if the points are scattered widely across the graph, this suggests a weak or no correlation between the variables.

Scatter plots can be enhanced by adding additional features such as trend lines, error bars, or color-coded markers to represent additional variables. 
They are commonly used in various fields, such as social sciences, finance, engineering, and biology, to visualize and explore relationships between variables.'''

import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

```

```python
#Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
'''Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])'''

#Ans-

'''The subplot() function in Matplotlib is used to create multiple plots or subplots in a single figure. 
It takes three arguments: the number of rows of subplots, the number of columns of subplots, and the index of the subplot to create.'''

import matplotlib.pyplot as plt
import numpy as np

# Data for the four lines
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Plot each line on a separate subplot
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Add titles to each subplot
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

# Add an overall title to the figure
fig.suptitle('Four Line Plots')

# Display the plots
plt.show()
```

```python
#Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
'''import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])'''

#Ans-

'''A bar plot is a type of graph that displays categorical data with rectangular bars, with the height or length of each bar proportional to the value it represents. 
It is often used to compare the values of different categories or groups.

A bar plot can be vertical or horizontal, depending on the orientation of the bars. In a vertical bar plot, the bars are plotted along the y-axis, with the category labels along the x-axis. 
In a horizontal bar plot, the bars are plotted along the x-axis, with the category labels along the y-axis.'''

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company, profit)
```

```python
# Add title and axis labels
plt.title('Profit by Company')
plt.xlabel('Company')
plt.ylabel('Profit')

# Display the plot
plt.show()

# Create a horizontal bar plot
plt.barh(company, profit)

# Add title and axis labels
plt.title('Profit by Company')
plt.xlabel('Profit')
plt.ylabel('Company')

# Display the plot
plt.show()
```
